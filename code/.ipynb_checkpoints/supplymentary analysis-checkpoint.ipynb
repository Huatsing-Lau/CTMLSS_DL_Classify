## DCA

# confidence intervals
95% CI for AUC-ROC and PRC-AUC

```python
from sklearn.utils import resample 
from sklearn.metrics import roc_auc_score, average_precision_score
def get_rocauc_95CI(y_true:List, y_score:List):
    """
    get CI by boostrapping.
    """
    n_samples = max(100,len(y_true))
    aucs = []
    for i in range(200):
        bootstrapSamples = resample(np.arange(0,len(y_true)), n_samples=n_samples, replace=True)  
        auc = roc_auc_score([y_true[i] for i in bootstrapSamples], [y_score[i] for i in bootstrapSamples])
        aucs.append(auc)
    aucs = sorted(aucs)
    ci = (aucs[4],aucs[-5])
    auc = roc_auc_score(y_true,y_score)
    return ci, auc


def get_ap_95CI(y_true:List, y_score:List):
    """
    get CI by boostrapping.
    """
    n_samples = max(100,len(y_true))
    aps = []
    for i in range(200):
        bootstrapSamples = resample(np.arange(0,len(y_true)), n_samples=n_samples, replace=True)  
        ap = average_precision_score([y_true[i] for i in bootstrapSamples], [y_score[i] for i in bootstrapSamples])
        aps.append(ap)
    aps = sorted(aps)
    ci = (aps[4],aps[-5])
    ap = roc_auc_score(y_true,y_score)
    return ci, ap
```

```python
feature_select_methods = ['MIC','RFE','EmbeddingLSVC','EmbeddingLR','EmbeddingTree','EmbeddingRF']
clf_names = ['Logistic', 'LDA', 'SVM', 'KNN', 'GaussianNB', 'DecisionTree',\
             'ExtraTrees', 'RandomForest', 'Bagging', 'AdaBoost', 'GradientBoost', 'MLP']

y_true = y_test.values.squeeze()
CI_rocauc = {method:{} for method in feature_select_methods}
CI_prauc = {method:{} for method in feature_select_methods}
for method in tqdm(feature_select_methods):
    for clf_name in clf_names:
        clf = all_result_dict[method][clf_name]['classifier']
        selected_features = all_result_dict[method]['selected_features']
        y_score = clf.predict_proba(X_test[selected_features].values)[:,1]
        # auc_roc
        ci, _ = get_rocauc_95CI(y_true,y_score)
        CI_rocauc[method][clf_name] = f'{ci[0]:.3f}~{ci[1]:.3f}'
        # prc_auc
        ci, _ = get_ap_95CI(y_true, y_score)
        CI_prauc[method][clf_name] = f'{ci[0]:.3f}~{ci[1]:.3f}'
        
display(pd.DataFrame(CI_rocauc))
display(pd.DataFrame(CI_prauc))

fn = os.path.join(dir_result,'测试集上的ROCAUC 95% CI.csv')
pd.DataFrame(CI_rocauc).to_csv(fn)

fn = os.path.join(dir_result,'测试集上的PRAUC 95% CI.csv')
pd.DataFrame(CI_prauc).to_csv(fn)
```
